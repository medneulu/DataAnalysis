```python
# Select only numerical columns for correlation calculation.
numerical_cols = sales.select_dtypes(include=np.number).columns # Selects only numeric columns
corr = sales[numerical_cols].corr() # Calculates correlation matrix for numeric columns
corr # Displays the correlation matrix
```

```python
fig = plt.figure(figsize=(8,8)) # Creates a new figure with specified size
# Convert the correlation matrix to a NumPy array to ensure it's numerical
corr_array = corr.to_numpy() # Converts the correlation DataFrame to a NumPy array
# The 'fignum' argument expects a single integer, not an array.
# Access the figure number using 'fig.number', which should be an integer.
# Remove the fignum argument as it's causing the issue.
plt.matshow(corr_array, cmap='RdBu') # Displays the correlation matrix as a heatmap
plt.xticks(range(len(corr.columns)), corr.columns, rotation='vertical') # Sets x-axis labels
plt.yticks(range(len(corr.columns)), corr.columns)  # Sets y-axis labels
```

```python
sales['Revenue_per_Age'] = sales['Revenue'] / sales['Customer_Age'] # Creates a new column for revenue per age

sales['Revenue_per_Age'].head() # Displays the first 5 values of the new column
```

```python
sales['Calculated_Cost'] = sales['Order_Quantity'] * sales['Unit_Cost'] # Calculates total cost per order

sales['Calculated_Cost'].head() # Displays first 5 calculated costs
```

```python
(sales['Calculated_Cost'] != sales['Cost']).sum() # Counts how many rows have mismatched calculated and given cost
```

```python
sales.plot(kind='scatter', x='Calculated_Cost', y='Profit', figsize=(6,6)) # Scatter plot of cost vs. profit
```

```python
sales['Calculated_Revenue'] = sales['Cost'] + sales['Profit'] # Calculates revenue from cost and profit

sales['Calculated_Revenue'].head() # Displays the first 5 calculated revenue values
```

```python
(sales['Calculated_Revenue'] != sales['Revenue']).sum() # Counts mismatches between calculated and actual revenue
```
